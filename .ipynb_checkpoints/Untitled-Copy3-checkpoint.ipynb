```python
X_train, X_test, y_train_3, y_test_3= train_test_split(X_train, y_train_3, test_size=0.25, random_state=0)
```

```python
logisticRegr = LogisticRegression()
```

```python
score = logisticRegr.score(X_test, y_test_3)
print(score)
```

```python
predictions = logisticRegr.predict(X_test)
```
