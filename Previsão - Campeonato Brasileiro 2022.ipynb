O modelo foi construído através de Regressão Logística. O dataset original contém 320 data points distribuídos ao longo de 14 features. Treze delas foram utilizadas na implementação modelo.

```python
import pandas as pd
import numpy as np
```

```python
cl_br = pd.read_csv('class_br.csv')
cl_br.head()
```

```python
# Features e Targets

X = cl_br.drop(['Clube','campeão','rebaixou','Liberta'],axis=1)
y_c = cl_br['campeão']
y_r = cl_br['rebaixou']
y_l = cl_br['Liberta']
```

```python
X.columns
```

```python
y_l.head()
```

```python
# Split train and test data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train_c, y_test_c = train_test_split(X, y_c, random_state = 3, test_size=0.25)
X_train, X_test, y_train_r, y_test_r = train_test_split(X, y_r, random_state = 3, test_size=0.25)
X_train, X_test, y_train_l, y_test_l = train_test_split(X, y_l, random_state = 3, test_size=0.25)
```

```python
# Import logreg function
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
param_grid = [
    {'penalty':['l1','l2'],
    'C': np.logspace(-3, 3, 20),
    'solver': ['lbfgs', 'newton-cg', 'liblinear', 'sag', 'saga'],
    'max_iter': [100,  1000, 2500, 5000]}
]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Optimize logreg parameters - 10-fol cross validation
clf_r = GridSearchCV(estimator = logreg, param_grid = param_grid, cv = 10, verbose = True, n_jobs= -1)
clf_l = GridSearchCV(estimator = logreg, param_grid = param_grid, cv = 10, verbose = True, n_jobs= -1)
clf_c = GridSearchCV(estimator = logreg, param_grid = param_grid, cv = 10, verbose = True, n_jobs= -1)
```

```python
# Libertadores
rl_l = clf_l.fit(X_train, y_train_l)
```

```python
# Rebaixamento
rl_r = clf_r.fit(X_train, y_train_r)
```

```python
# Campeão
rl_c = clf_c.fit(X_train, y_train_c)
```

```python
# Accurácia
rl_c.best_score_
```

```python
rl_c.score(X_test, y_test_c)
```

# A seguir, apenas algumas análises internas que fiz por curiosidade

Implementação de bootstraps para identificar o intervalo de confiança de algumas métricas.

```python
# Função para bootstrap + gráfico
def bootstrap(C, t, i):
    C = C.astype(float)
    dis = np.zeros(i)
    ind = np.arange(len(C))
    for i in range(i):
        random.shuffle(ind)
        for v in range(t):
            w = ind[v]
            dis[i] += C[w]
        dis[i] = dis[i]/t
    #print(dis)
    return plt.figure(figsize=(10,5)), sns.displot(dis, fill = True, bins=25), pd.DataFrame(dis).describe(percentiles=[.025, .5, .975])
```

```python
# Intervalo de confiança para classificados a libertadores
bt = comp[comp['Liberta'] == 1].reset_index().drop('index', axis=1)
aprov = bt['aprov']
bootstrap(aprov, 50, 100_000)
```

```python
bt = comp[comp['Liberta'] == 1].reset_index().drop('index', axis=1)
aprov = bt['V']
bootstrap(aprov, 50, 200_000)
```

```python
bt = comp[comp['rebaixou'] == 1].reset_index().drop('index', axis=1)
aprov = bt['V']
bootstrap(aprov, 20, 200_000)
```

```python
# Intervalo de confiança para os rebaixados
bt = comp[comp['rebaixou'] == 1].reset_index().drop('index', axis=1)
aprov = bt['aprov']
bootstrap(aprov, 30, 100_000)
```
